```python
import folium
from folium import DivIcon

# Initialize the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Group outcomes by launch site and class (0/1)
site_outcome = spacex_df.groupby('Launch Site')['class'].value_counts().reset_index(name='count')

# Loop through launch sites with their coordinates
for index, row in launch_sites_df.iterrows():
    site_name = row['Launch Site']
    lat = row['Lat']
    lon = row['Long']
    
    # Get number of successes and failures for this site
    successes = site_outcome[(site_outcome['Launch Site'] == site_name) & (site_outcome['class'] == 1)]['count']
    failures = site_outcome[(site_outcome['Launch Site'] == site_name) & (site_outcome['class'] == 0)]['count']
    
    # Handle cases where a site may not have any failures or successes
    success_count = int(successes.values[0]) if not successes.empty else 0
    failure_count = int(failures.values[0]) if not failures.empty else 0

    # Create popup text
    popup_text = f"<b>{site_name}</b><br>Successes: {success_count}<br> Failures: {failure_count}"

    # Add circle
    folium.Circle(
        location=(lat, lon),
        radius=1000,
        color='#000000',
        fill=True
    ).add_child(folium.Popup(popup_text)).add_to(site_map)
    
    # Add marker label
    folium.map.Marker(
        location=(lat, lon),
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{site_name}</b></div>'
        )
    ).add_to(site_map)

# Show map
site_map
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Apply a function to determine the marker color
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')
spacex_df.head(1)
```

```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# for each row in spacex_df data frame
# create a Marker object with its coordinate
# and customize the Marker's icon property to indicate if this launch was successed or failed, 
# e.g., icon=folium.Icon(color='white', icon_color=row['marker_color']
for index, row in spacex_df.iterrows():
    lat = row['Lat']
    lon = row['Long']
    site = row['Launch Site']
    outcome = row['class']
    icon_color = 'green' if outcome == 1 else 'red'
    
    marker = folium.Marker(
        location=(lat, lon),
        popup=f"{site} - {'Success' if outcome == 1 else 'Failure'}",
        icon=folium.Icon(color='white', icon_color=icon_color, icon='info-sign')
    )
    
    marker_cluster.add_child(marker)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
